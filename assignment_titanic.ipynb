```python
#filling the missing values

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(method = 'ffill')
df['Cabin'] = df['Cabin'].fillna(method = 'ffill')
```

```python
# 5.checking outlier

num_col = ['Fare','Age','Survived','Pclass']
```

```python
for i in num_col:
    plt.figure()
    plt.boxplot(df[i])
    plt.title(i)
```

```python
#it is found that 'age' & 'fare' has large number of outliers,so for removing them


Q1 = np.percentile(df['Fare'],25,method = 'midpoint')
Q2 = np.percentile(df['Fare'],50,method = 'midpoint')
Q3 = np.percentile(df['Fare'],75,method = 'midpoint')

print("q1 is",Q1)
print("q2 is",Q2)
print("q3 is",Q3)
```

```python
IQR = Q3-Q1
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print('IQR is',IQR)
print('low_lim is',low_lim)
print('up_lim is',up_lim)
```

```python
#to catch outliers

outlier = []
for x in df ["Fare"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
```

```python
outlier
```

```python
#remove these outliers from data

removed = df["Fare"]>up_lim
df.loc[removed].index
```

```python
df.drop(df.loc[removed].index,inplace = True)
```

```python
#outlier in age

Q1 = np.percentile(df['Age'],25,method = 'midpoint')
Q2 = np.percentile(df['Age'],50,method = 'midpoint')
Q3 = np.percentile(df['Age'],75,method = 'midpoint')

print("q1 is",Q1)
print("q2 is",Q2)
print("q3 is",Q3)
```

```python
IQR = Q3-Q1
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print('IQR is',IQR)
print('low_lim is',low_lim)
print('up_lim is',up_lim)
```

```python
outlier = []
for x in df ["Age"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
```

```python
# 6. min max scaling on the feature set

x=df.drop(['Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)
```

```python
df= pd.get_dummies(df)
```

```python
y=df['Survived']
x=df.drop('Survived',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
lr=linear_model.LinearRegression()
model=lr.fit(x_train, y_train)
from sklearn.metrics import mean_squared_error
pred=model.predict(x_test)
print('MSE is ', mean_squared_error(y_test, pred))
from sklearn.metrics import r2_score
print('R2 score is ', r2_score(y_test, pred))
```

```python
from sklearn import preprocessing
min_max = preprocessing.MinMaxScaler(feature_range=(0,2))
x = min_max.fit_transform(x)
print(x)
x=pd.DataFrame(x)
x.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
lr=linear_model.LinearRegression()
model=lr.fit(x_train, y_train)
from sklearn.metrics import mean_squared_error
pred=model.predict(x_test)
print('MSE is ', mean_squared_error(y_test, pred))
from sklearn.metrics import r2_score
print('R2 score is ', r2_score(y_test, pred))
```
